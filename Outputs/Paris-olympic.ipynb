```python
athletes = spark.read.format("parquet").load("wasbs://paris-olympic-data@pocstorageretail.blob.core.windows.net/raw-data/athletes.csv")
coaches  = spark.read.format("parquet").load("wasbs://paris-olympic-data@pocstorageretail.blob.core.windows.net/raw-data/coaches.csv")
EntriesGender = spark.read.format("csv").option("header","true").option("inferSchema","true").load("wasbs://paris-olympic-data@pocstorageretail.blob.core.windows.net/raw-data/EntriesGender.csv")
medals = spark.read.format("csv").option("header","true").option("inferSchema","true").load("wasbs://paris-olympic-data@pocstorageretail.blob.core.windows.net/raw-data/medals.csv")
teams = spark.read.format("csv").option("header","true").option("inferSchema","true").load("wasbs://paris-olympic-data@pocstorageretail.blob.core.windows.net/raw-data/teams.csv")


```

```python
EntriesGender = EntriesGender.withColumn("female",EntriesGender["Female"].cast("int"))
EntriesGender = EntriesGender.withColumn("male",EntriesGender["Male"].cast("int"))
EntriesGender = EntriesGender.withColumn("Grand Total",EntriesGender["Grand Total"].cast("int"))
```

```python
#find the Top countries with the Highest Number of gold medals

top_gold_medals = medals.orderBy("Gold Medal",ascending=False).select("Row Labels","Gold Medal").limit(10)
top_gold_medals.show()

```

```python
#Calculate the avg number of Entries by gender for Each Discipline

avg_entries_by_gender = EntriesGender.withColumn("Avg_Female",EntriesGender["female"]/EntriesGender["Grand Total"]).withColumn("Avg_Male",EntriesGender["male"]/EntriesGender["Grand Total"])
avg_entries_by_gender.show()


 
 
```
